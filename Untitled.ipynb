## ScanCount & VisitNumber

### 반품

ScanCount는 상품을 몇개 샀는지를 나타낸다. 이때 - 수는 반품했음을 의미한다.     

우리가 찾고자 하는 것은 구매자의 요일, 구매상품 등의 정보를 이용해서 이 사람의 소비 유형을 파악하는 것. 그런데 소비 유형은 매번 달라진다. 어제 왔던 사람이 오늘 또 왔다고 해서 똑같은 소비 유형을 보이지 않을 수 있다. 매번 방문 목적이 다를 수 있다. 이에 visitnumber가 다르면 아예 다른 사람이라고 봐야한다고 생각.      
       
이때, 어떤 상품을 반품한다는 정보는 이 사람의 소비 유형을 보여주진 않는다. 어떤 상품을 반품한다는 뜻은 과거에 그 상품을 구매했다는 뜻이고, 과거에 구매한 정보는 다른 visitnumber사람이 가지고 있으니, 현재의 나의 소비와는 상관이 없다. 
            
이때 혹시 반품만 하러 온 사람을 분류하는 type이 있을 수 있으니 먼저 확인해보기.


엑셀로 확인해보니, 반품만 하러 온 사람도 999라는 type으로 분류되고,    
그날 와서 사고 판매하거나, 판매하고 사가서 총 count가 0인 사람들도 999라는 type으로 분류된다. 즉 산만큼 반품하고 간 사람들도 반품만 하러 온 사람들이라고 분류하자.

```python
df[df['TripType']==999].head()
```

### 소비자 vs 생산자?

먼저 들었던 생각은, 크게 소비자와 생산자는 구매 수량이 다를 테니, 구매수량을 보고 구매자와 생산자를 나눌 수 있지 않을까?

```python
def evaluate(X_train,X_test,y_train,y_test):
    # Name of models
    model_name_list=['Logistic','LDA','Naive','RandomForest','GBT']
    
    model1 = LogisticRegression()
    model2 = LinearDiscriminantAnalysis()
    model3 = GaussianNB()
    model4 = RandomForestClassifier(n_estimators=100,max_depth=8,random_state=42)
    model5 = GradientBoostingClassifier(n_estimators=100,max_depth=8,random_state=42)
    
    
    # DataFrame for result
    result = pd.DataFrame()
     
    # Train and predict with each model
    accuracy=[]
        
    for i, model in enumerate([model1,model2,model3,model4,model5]):
        accuracy.append(cross_val_score(model,X_train,y_train,cv=5).mean())
     
    result['model'] = ['Logistic','LDA','Naive','RandomForest','GBT']
    result['accuracy'] = accuracy
    
    
    return result
```
